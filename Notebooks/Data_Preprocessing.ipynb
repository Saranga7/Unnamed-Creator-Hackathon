```python
def remove_rupees(string):
    return string[1:]

def remove_commas(string):
    string = string.replace(',', '').replace('or', str(np.nan))
    return string
```

```python
df.iloc[:, 3:5] = df.iloc[:, 3:5].applymap(remove_rupees)
df.iloc[:, 3:5] = df.iloc[:, 3:5].applymap(remove_commas)
```

```python
df.iloc[:, 5] = pd.to_numeric(df.iloc[:, 5], errors = 'coerce')
```

```python
for i in range(3, 8):
    df.iloc[:, i] = df.iloc[:, i].astype('float')
```

# Missing values handling for Test Set, same procedure is followed for Training Set too
#### We update the missing values of each column w.r.t 'Location' column, i.e. we take the average of all the restaurants in that specific area, assuming the delivery_time would depend on the Location

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Location'].value_counts()
```

```python
df['Votes'] = df.groupby('Location')['Votes'].transform(
    lambda grp: grp.fillna(np.mean(grp))
)
```

```python
df['Reviews'] = df.groupby('Location')['Reviews'].transform(
    lambda grp: grp.fillna(np.mean(grp))
)
```

```python
df['Average_Cost'] = df.groupby('Location')['Average_Cost'].transform(
    lambda grp: grp.fillna(np.mean(grp))
)
```

```python
df.isnull().sum()
```

```python
df.to_csv('test_dataset.csv', index = False, header = True)
```
